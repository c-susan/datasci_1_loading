**CSV 1:** This dataset contains information on Medicare providers specializing in Home Infusion Therapy. Columns in the dataset include the following: Enrollment ID, Legal Business Name, Street Address Line 1, Street Address Line 2, City, State, Zip Code, Practice Location Phone Number, Specialty Name, Geographic Location Type Description, Geographic Location City Name, Geographic Location State Code, Geographic Location Zip Code, and State County Name. The column Enrollment ID indicates the providers' Medicaid enrollment ID, and Legal Business Name indicates the providers' clinic's name. The Specialty Name column indicates the specializations of the providers, which is Home Infusion Therapy. Phone numbers of each provider are listed in the Practice Location Phone Number column. The remaining columns indicate the providers' addresses, including street address, zip code, city, state, county, and geographic location description.
